```python
import mdtraj as md
import numpy as np
```

```python
t1 = md.load('C:/Users/PC/Dropbox/hbonds/trajectory/f2/cco_pr_popc_7ns_center.xtc',\
             top='C:/Users/PC/Dropbox/hbonds/trajectory/f2/cco_pr_popc_0ns.gro')
t2 = md.load('C:/Users/PC/Dropbox/hbonds/trajectory/f2/cco_pr_popc_14ns_center.xtc',\
            top='C:/Users/PC/Dropbox/hbonds/trajectory/f2/cco_pr_popc_0ns.gro')
t3 = md.load('C:/Users/PC/Dropbox/hbonds/trajectory/f2/cco_pr_popc_26ns_center.xtc',\
            top='C:/Users/PC/Dropbox/hbonds/trajectory/f2/cco_pr_popc_0ns.gro')
t4 = md.load('C:/Users/PC/Dropbox/hbonds/trajectory/f2/cco_pr_popc_38ns_center.xtc',\
            top='C:/Users/PC/Dropbox/hbonds/trajectory/f2/cco_pr_popc_0ns.gro')
t5 = md.load('C:/Users/PC/Dropbox/hbonds/trajectory/f2/cco_pr_popc_50ns_center.xtc',\
            top='C:/Users/PC/Dropbox/hbonds/trajectory/f2/cco_pr_popc_0ns.gro')

t=t1.join(t2, check_topology=True, discard_overlapping_frames=True)
t=t.join(t3, check_topology=True, discard_overlapping_frames=True)
t=t.join(t4, check_topology=True, discard_overlapping_frames=True)
t=t.join(t5, check_topology=True, discard_overlapping_frames=True)

```

```python
print(t.time)
```

```python
selection1 = t.topology.select_expression('resn leu ')
print(selection1)
```

```python
"""
all_glu_dists: Glu286 CD index [-6] 2214, PRD of heme a3 CAD(-5)
all_trp172_dists: WRP172 NE1 index, PRD of heme a3 CAD(-5)
leu255_ile99_dists: LEU255 CG index, ILE99 CB index
"""
### all_glu_dists stores the distance between glu286 and PRD of heme a3


#select residue atoms 
glu_286=[atom for atom in t.topology.atoms if ( (atom.residue.index == 274))]
he3=[atom for atom in t.topology.atoms if (atom.residue.index == 0)]
hea=[atom for atom in t.topology.atoms if (atom.residue.index == 1)]
trp_172=[atom for atom in t.topology.atoms if ( (atom.residue.index == 160))]
ile_99=[atom for atom in t.topology.atoms if ( (atom.residue.index == 87))]
leu_255=[atom for atom in t.topology.atoms if ( str(atom)[0:6] == "LEU255")]

#select certain atom in residue: glu286, heme a3, ile99, leu255 
glu_286_CD=glu_286[-6]
trp_172_NE1=trp_172[10]
prda3_CAD=he3[-9]
leu_255_CG=leu_255[7]
ile_99_CB=ile_99[4]

##compute distances 
glu_286_prda3_pair= np.asarray([[glu_286_CD.index, prda3_CAD.index]])
trp_172_prda3_pair= np.asarray([[trp_172_NE1.index,prda3_CAD.index]])
leu_255_ile_99_pair= np.asarray([[leu_255_CG.index,ile_99_CB.index]])

glu_286_prda3_dist=md.compute_distances(t, glu_286_prda3_pair, periodic=False)
trp_172_prda3_dist=md.compute_distances(t, trp_172_prda3_pair, periodic=False)
leu_255_ile_99_dist=md.compute_distances(t, leu_255_ile_99_pair, periodic=False)
```
