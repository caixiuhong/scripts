```python
#create lists for pdb file
all_files=[]
all_subdirs=[]
substates=[]

for root,dirs,files in os.walk(inputpathway):
    if not dirs:
        subdirpath=os.path.join(root)
        all_subdirs.append(subdirpath)
        #outputsubdir=re.sub(inputpathway,outputpathway,subdirpath) 
        #os.mkdir(outputsubdir)
        substate=os.path.split(os.path.split(root)[-2])[-1]+'_'+os.path.split(root)[-1]
        substates.append(substate)
        
    for name in files:
        inputfilepath=os.path.join(root,name)
        #print("creating %s." % (str(inputfilepath)))
        all_files.append(inputfilepath)

#print (all_subdirs)
print (substates)

pdbs=[]
time_points=[]
for index, subdir in enumerate(all_subdirs):
    pdb=[]
    time_point=[]
    for f in os.listdir(subdir):
        pdb.append(os.path.join(subdir, f))
        time_point.append(f[-12:-10])
    pdbs.append(pdb)
    time_points.append(time_point)
#print pdbs
print (time_points)

```

```python
### all_water_numbers store the water numbers within 3.5 A of Thr100, Asn96(O, N) or Trp95

query_atom_ids = [807, 777, 778, 764]
all_water_numbers = []
for substate_index, substate in enumerate(pdbs):
    print (substates[substate_index])
    all_water_number=[]
    for index, pdb in enumerate(substate):
        print ("Processing: ", pdb)
        snapshot = md.load_pdb(pdb)    
        q = query_atom_ids
        print ([snapshot.topology.atom(index).residue for index in q])
        wat_oxygen_atom_ids = snapshot.topology.select("water and name O")
        cav_waters = md.compute_neighbors(
                snapshot, 0.35, q,
                haystack_indices=wat_oxygen_atom_ids)
        print (cav_waters)
        #print cav_waters[0].shape[0]
        all_water_number.append(cav_waters[0].shape[0])
        #print all_water_number
    all_water_numbers.append(all_water_number)
    
print (all_water_numbers)
```

```python
print ("Number of waters in the Thr100")
for index, substate in enumerate(substates):
    print (index,substate)
    print ("Substate: "+substate)
    for index_time, time_point in enumerate(time_points[index]):
        print (time_point, all_water_numbers[index][index_time])
```

```python
### all_water_numbers_e286 stores the water numbers within 4.5 A of Glu286
all_water_numbers_e286 = []
for substate_index, substate in enumerate(pdbs):
    print (substates[substate_index])
    all_water_number_e286=[]
    for index, pdb in enumerate(substate):
        print ("Processing: ", pdb)
        snapshot = md.load_pdb(pdb)    
        all_res = [residue for residue in snapshot.topology.residues]
        all_glus = snapshot.top.select("resn GLU")
        glu286_at_ids = [at for at in all_glus if str(snapshot.topology.atom(at))[0:6] == "GLU286"]
        glu286_at_names = [str(snapshot.topology.atom(at)) for at in glu286_at_ids]
        print (glu286_at_names)
        glu286_OE_at_ids = np.asarray([glu286_at_ids[i] for i in range(len(glu286_at_names)) if "OE" in glu286_at_names[i]]) 
        print (glu286_OE_at_ids)
        wat_oxygen_atom_ids = snapshot.topology.select("water and name O")
        glu286_waters = md.compute_neighbors(
                snapshot, 0.45, glu286_OE_at_ids,
                haystack_indices=wat_oxygen_atom_ids)
        print (glu286_waters)
        all_water_number_e286.append(glu286_waters[0].shape[0])
    all_water_numbers_e286.append(all_water_number_e286)
    
print (all_water_numbers_e286)
```

```python
### all_thr_dists stores the distance between glu286 and PRD of heme a
### all_thr_dists_2 stores the distance between glu286 and PRD of heme a3

query_atom_ids_2 = [807, 777, 764]
all_thr_dists = []
all_asn_dists=[]
all_trp_dists=[]
all_thr_dists_2 = []
all_asn_dists_2=[]
all_trp_dists_2=[]

for substate_index, substate in enumerate(pdbs):
    #print substates[substate_index]
    all_thr_dist=[]
    all_asn_dist=[]
    all_trp_dist=[]
    all_thr_dist_2=[]
    all_asn_dist_2=[]
    all_trp_dist_2=[]
    
    for index, pdb in enumerate(substate):
        #print "Processing: ", pdb
        snapshot = md.load_pdb(pdb)    
        q = query_atom_ids_2
        #print [snapshot.topology.atom(index) for index in q]
        hea = snapshot.top.select("resn HEA")
        hea_at_names = [str(snapshot.topology.atom(at)) for at in hea]
        #print hea_at_names
        at_indices_thr = np.asarray([[q[0], hea[-5]]])
        at_indices_asn = np.asarray([[q[1], hea[-5]]])
        at_indices_trp = np.asarray([[q[2], hea[-5]]])
        thr_dist = md.compute_distances(snapshot, at_indices_thr, periodic=False)
        asn_dist = md.compute_distances(snapshot, at_indices_asn, periodic=False)
        trp_dist = md.compute_distances(snapshot, at_indices_trp, periodic=False)
        all_thr_dist.append(thr_dist[0][0])
        all_asn_dist.append(asn_dist[0][0])
        all_trp_dist.append(trp_dist[0][0])
        
        at_indices_thr_2 = np.asarray([[q[0], he3[-5]]])
        at_indices_asn_2 = np.asarray([[q[1], he3[-5]]])
        at_indices_trp_2 = np.asarray([[q[2], he3[-5]]])
        thr_dist_2 = md.compute_distances(snapshot, at_indices_thr_2, periodic=False)
        asn_dist_2 = md.compute_distances(snapshot, at_indices_asn_2, periodic=False)
        trp_dist_2 = md.compute_distances(snapshot, at_indices_trp_2, periodic=False)
        all_thr_dist_2.append(thr_dist_2[0][0])
        all_asn_dist_2.append(asn_dist_2[0][0])
        all_trp_dist_2.append(trp_dist_2[0][0])
        
        
    all_thr_dists.append(all_thr_dist)
    all_asn_dists.append(all_asn_dist)
    all_trp_dists.append(all_trp_dist)
    
    all_thr_dists_2.append(all_thr_dist_2)
    all_asn_dists_2.append(all_asn_dist_2)
    all_trp_dists_2.append(all_trp_dist_2)

print (all_thr_dists)
print (all_asn_dists)
print (all_trp_dists)

print (all_thr_dists_2)
print (all_asn_dists_2)
print (all_trp_dists_2)
```

```python
"""
CU-CU distance at CUA

"""

cua_cua_dists = []


for substate_index, substate in enumerate(pdbs):
    
    cua_cua_dist=[]
    
    
    
    for index, pdb in enumerate(substate):
        #print ("Processing: "+ pdb)
        snapshot = md.load_pdb(pdb)  
        all_res = [residue for residue in snapshot.topology.residues]
        
        all_cus=snapshot.top.select("resn CUA")
        
       
        
        cua_at_ids =[at for at in all_cus if str(snapshot.topology.atom(at))[0:3] == "CUA"]
        
        cua_at_names = [str(snapshot.topology.atom(at)) for at in cua_at_ids]
        

        at_indices_cua_cua = np.asarray([[cua_at_ids[0], cua_at_ids[1]]])
        cua_cua = md.compute_distances(snapshot, at_indices_cua_cua, periodic=True)
        cua_cua_dist.append(cua_cua[0][0])
        
        
    cua_cua_dists.append(cua_cua_dist)
    
print(cua_cua_dists)
```
