```python
#p4 trajectory
dpdro4_t = md.load('/Users/caixiuhong/Dropbox/cai/btype_cco/snapshot/Btype_CcO_MD/Btype_MD/4_dpdro3/md_nvt_25ns_100.dcd',\
             top='/Users/caixiuhong/Dropbox/cai/btype_cco/snapshot/Btype_CcO_MD/Btype_MD/4_dpdro3/md_nvt_init.pdb')
```

```python
##atom selection: his376 atom NE1, ND1; asp372 atom OD1, OD2; heme a3 atom O1A, O2A
trajs=[dddro1_t, pddro2_t, ppdro3_t, dpdro4_t]

his376_NE2={}; his376_ND1={}; asp372_OD1={}; asp372_OD2={}; he3_O1A={}; he3_O2A={}

for index, traj in enumerate(trajs):
    his376_NE2[index]=[atom.index for atom in traj.topology.atoms if ((atom.residue.name == 'HIS') and (atom.residue.index == 369) \
                                                                  and (atom.name =='NE2') )]
    his376_ND1[index]=[atom.index for atom in traj.topology.atoms if ((atom.residue.name == 'HIS') and (atom.residue.index == 369) \
                                                                  and (atom.name =='ND1') )]
    asp372_OD1[index]=[atom.index for atom in traj.topology.atoms if ((atom.residue.name == 'ASP') and (atom.residue.index == 365)\
                                                               and (atom.name == 'OD1'))]
    asp372_OD2[index]=[atom.index for atom in traj.topology.atoms if ((atom.residue.name == 'ASP') and (atom.residue.index == 365)\
                                                               and (atom.name == 'OD2'))]
    
    he3_O1A[index]=[atom.index for atom in traj.topology.atoms if ((atom.residue.name == 'HEM') and (atom.residue.index == 0) \
                                                            and (atom.name == 'O1A'))]
    he3_O2A[index]=[atom.index for atom in traj.topology.atoms if ((atom.residue.name == 'HEM') and (atom.residue.index == 0) \
                                                            and (atom.name == 'O2A'))]
    
print(his376_NE2, his376_ND1, asp372_OD1, asp372_OD2, he3_O1A, he3_O2A)

```

```python
##compute distances 

his376_NE2_he3_O1A={}; his376_NE2_he3_O1A_dist={}
his376_NE2_he3_O2A={}; his376_NE2_he3_O2A_dist={}
his376_ND1_he3_O1A={}; his376_ND1_he3_O1A_dist={}
his376_ND1_he3_O2A={}; his376_ND1_he3_O2A_dist={}

asp372_OD1_he3_O1A={}; asp372_OD1_he3_O1A_dist={}
asp372_OD1_he3_O2A={}; asp372_OD1_he3_O2A_dist={}
asp372_OD2_he3_O1A={}; asp372_OD2_he3_O1A_dist={}
asp372_OD2_he3_O2A={}; asp372_OD2_he3_O2A_dist={}

his376_NE2_asp372_OD1={}; his376_NE2_asp372_OD1_dist={}
his376_NE2_asp372_OD2={}; his376_NE2_asp372_OD2_dist={}
his376_ND1_asp372_OD1={}; his376_ND1_asp372_OD1_dist={}
his376_ND1_asp372_OD2={}; his376_ND1_asp372_OD2_dist={}


for index, traj in enumerate(trajs):
    
    #his376 and he3 pra distance
    his376_NE2_he3_O1A[index]=np.asarray([[his376_NE2[index][0], he3_O1A[index][0]]])
    his376_NE2_he3_O1A_dist[index]=md.compute_distances(traj, his376_NE2_he3_O1A[index], periodic=True)
    his376_NE2_he3_O2A[index]=np.asarray([[his376_NE2[index][0], he3_O2A[index][0]]])
    his376_NE2_he3_O2A_dist[index]=md.compute_distances(traj, his376_NE2_he3_O2A[index], periodic=True)
    
    his376_ND1_he3_O1A[index]=np.asarray([[his376_ND1[index][0], he3_O1A[index][0]]])
    his376_ND1_he3_O1A_dist[index]=md.compute_distances(traj, his376_ND1_he3_O1A[index], periodic=True)
    his376_ND1_he3_O2A[index]=np.asarray([[his376_ND1[index][0], he3_O2A[index][0]]])
    his376_ND1_he3_O2A_dist[index]=md.compute_distances(traj, his376_ND1_he3_O2A[index], periodic=True)
    
    #asp372 and he3 pra distance
    asp372_OD1_he3_O1A[index]=np.asarray([[asp372_OD1[index][0], he3_O1A[index][0]]])
    asp372_OD1_he3_O1A_dist[index]=md.compute_distances(traj, asp372_OD1_he3_O1A[index], periodic=True)
    asp372_OD1_he3_O2A[index]=np.asarray([[asp372_OD1[index][0], he3_O2A[index][0]]])
    asp372_OD1_he3_O2A_dist[index]=md.compute_distances(traj, asp372_OD1_he3_O2A[index], periodic=True)
    
    asp372_OD2_he3_O1A[index]=np.asarray([[asp372_OD2[index][0], he3_O1A[index][0]]])
    asp372_OD2_he3_O1A_dist[index]=md.compute_distances(traj, asp372_OD2_he3_O1A[index], periodic=True)
    asp372_OD2_he3_O2A[index]=np.asarray([[asp372_OD2[index][0], he3_O2A[index][0]]])
    asp372_OD2_he3_O2A_dist[index]=md.compute_distances(traj, asp372_OD2_he3_O2A[index], periodic=True)

    #his376 and asp372 distance
    his376_NE2_asp372_OD1[index]=np.asarray([[his376_NE2[index][0], asp372_OD1[index][0]]])
    his376_NE2_asp372_OD1_dist[index]=md.compute_distances(traj, his376_NE2_asp372_OD1[index], periodic=True)
    his376_NE2_asp372_OD2[index]=np.asarray([[his376_NE2[index][0], asp372_OD2[index][0]]])
    his376_NE2_asp372_OD2_dist[index]=md.compute_distances(traj, his376_NE2_asp372_OD2[index], periodic=True)
    
    his376_ND1_asp372_OD1[index]=np.asarray([[his376_ND1[index][0], asp372_OD1[index][0]]])
    his376_ND1_asp372_OD1_dist[index]=md.compute_distances(traj, his376_ND1_asp372_OD1[index], periodic=True)
    his376_ND1_asp372_OD2[index]=np.asarray([[his376_ND1[index][0], asp372_OD2[index][0]]])
    his376_ND1_asp372_OD2_dist[index]=md.compute_distances(traj, his376_ND1_asp372_OD2[index], periodic=True)
    
   
    
#print( his376_NE2_he3_O1A_dist)

```

```python
#get min distance between his376 and he3 pra, his376 and asp372, asp372 and he3 pra
min_his376_he3_dist={}
min_his376_NE_he3_dist={}
min_his376_ND_he3_dist={}
min_his376_asp372_dist={}
min_asp372_he3_dist={}
for index, traj in enumerate(trajs):
    min_his376_he3_dist[index]=np.minimum(his376_NE2_he3_O1A_dist[index], his376_NE2_he3_O2A_dist[index])
    min_his376_he3_dist[index]=np.minimum(min_his376_he3_dist[index], his376_ND1_he3_O1A_dist[index])
    min_his376_he3_dist[index]=np.minimum(min_his376_he3_dist[index], his376_ND1_he3_O2A_dist[index])

    min_his376_NE_he3_dist[index]=np.minimum(his376_NE2_he3_O1A_dist[index], his376_NE2_he3_O2A_dist[index]) 
    
    min_his376_ND_he3_dist[index]=np.minimum(his376_ND1_he3_O1A_dist[index], his376_ND1_he3_O2A_dist[index]) 
    
    min_his376_asp372_dist[index]=np.minimum( his376_NE2_asp372_OD1_dist[index], his376_NE2_asp372_OD2_dist[index])
    min_his376_asp372_dist[index]=np.minimum(min_his376_asp372_dist[index], his376_ND1_asp372_OD1_dist[index])
    min_his376_asp372_dist[index]=np.minimum(min_his376_asp372_dist[index], his376_ND1_asp372_OD2_dist[index])
    
    min_asp372_he3_dist[index]=np.minimum(asp372_OD1_he3_O1A_dist[index], asp372_OD1_he3_O2A_dist[index])
    min_asp372_he3_dist[index]=np.minimum(min_asp372_he3_dist[index], asp372_OD2_he3_O1A_dist[index])
    min_asp372_he3_dist[index]=np.minimum(min_asp372_he3_dist[index], asp372_OD2_he3_O2A_dist[index])

#print(min_his376_he3_dist[3][0],min_his376_asp372_dist[3][0],min_asp372_he3_dist[3][0])
```

```python
#define High distance seperation and low distance seperation of 
#his376_he3 dist, asp372_he3 dist, and his376_asp372 dist
h_line_min_his376_he3_dist=4.0
l_line_min_his376_he3_dist=3.0
h_line_min_his376_asp372_dist=4.5
l_line_min_his376_asp372_dist=3.5
h_line_min_asp372_he3_dist=4.0
l_line_min_asp372_he3_dist=3.0

LLL={}; HLL={};LHL={};LLH={};HHL={};HLH={};LHH={};HHH={}

#divide the 3D data into eight baskets [LLL], [HLL],[LHL],[LLH],[HHL],[HLH],[LHH],[HHH]
for index, traj in enumerate(trajs):
    
    LLL_list=[]; HLL_list=[];LHL_list=[];LLH_list=[];HHL_list=[];HLH_list=[];LHH_list=[];HHH_list=[]
    
    for i, x_dist in enumerate(min_his376_he3_dist[index]*10):
        if x_dist < l_line_min_his376_he3_dist and \
        (min_his376_asp372_dist[index]*10)[i] < l_line_min_his376_asp372_dist and \
        (min_asp372_he3_dist[index]*10)[i] < l_line_min_asp372_he3_dist:
            ##LLL[traj_index]:[n_frame, [min_his376_he3_dist, min_his376_asp372_dist,min_asp372_he3_dist]]
            LLL_list.append([i, x_dist,(min_his376_asp372_dist[index]*10)[i], (min_asp372_he3_dist[index]*10)[i]])
        elif x_dist > h_line_min_his376_he3_dist and \
        (min_his376_asp372_dist[index]*10)[i] < l_line_min_his376_asp372_dist and \
        (min_asp372_he3_dist[index]*10)[i] < l_line_min_asp372_he3_dist:
            HLL_list.append([i, x_dist,(min_his376_asp372_dist[index]*10)[i], (min_asp372_he3_dist[index]*10)[i]])
        elif x_dist < l_line_min_his376_he3_dist and \
        (min_his376_asp372_dist[index]*10)[i] > l_line_min_his376_asp372_dist and \
        (min_asp372_he3_dist[index]*10)[i] < l_line_min_asp372_he3_dist:
            LHL_list.append([i, x_dist,(min_his376_asp372_dist[index]*10)[i], (min_asp372_he3_dist[index]*10)[i]])
        elif x_dist < l_line_min_his376_he3_dist and \
        (min_his376_asp372_dist[index]*10)[i] < l_line_min_his376_asp372_dist and \
        (min_asp372_he3_dist[index]*10)[i] > h_line_min_asp372_he3_dist:
            ##LLL[traj_index]=[n_frame, [min_his376_he3_dist, min_his376_asp372_dist,min_asp372_he3_dist]]
            LLH_list.append([i, x_dist,(min_his376_asp372_dist[index]*10)[i], (min_asp372_he3_dist[index]*10)[i]])
        elif x_dist > h_line_min_his376_he3_dist and \
        (min_his376_asp372_dist[index]*10)[i] > h_line_min_his376_asp372_dist and \
        (min_asp372_he3_dist[index]*10)[i] < l_line_min_asp372_he3_dist:
            HHL_list.append([i, x_dist,(min_his376_asp372_dist[index]*10)[i], (min_asp372_he3_dist[index]*10)[i]])
        elif x_dist > h_line_min_his376_he3_dist and \
        (min_his376_asp372_dist[index]*10)[i] < l_line_min_his376_asp372_dist and \
        (min_asp372_he3_dist[index]*10)[i] > h_line_min_asp372_he3_dist:
            HLH_list.append([i, x_dist,(min_his376_asp372_dist[index]*10)[i], (min_asp372_he3_dist[index]*10)[i]])
        elif x_dist < l_line_min_his376_he3_dist and \
        (min_his376_asp372_dist[index]*10)[i] > h_line_min_his376_asp372_dist and \
        (min_asp372_he3_dist[index]*10)[i] > h_line_min_asp372_he3_dist:
            ##LLL[traj_index]:[n_frame, min_his376_he3_dist, min_his376_asp372_dist,min_asp372_he3_dist]
            LHH_list.append([i, x_dist,(min_his376_asp372_dist[index]*10)[i], (min_asp372_he3_dist[index]*10)[i]])
        elif x_dist > h_line_min_his376_he3_dist and \
        (min_his376_asp372_dist[index]*10)[i] > h_line_min_his376_asp372_dist and \
        (min_asp372_he3_dist[index]*10)[i] > h_line_min_asp372_he3_dist:
            ##LLL[traj_index]:[n_frame, [min_his376_he3_dist, min_his376_asp372_dist,min_asp372_he3_dist]]
            HHH_list.append([i, x_dist,(min_his376_asp372_dist[index]*10)[i], (min_asp372_he3_dist[index]*10)[i]])
        
    
    LLL[index]=LLL_list
    HLL[index]=HLL_list
    LHL[index]=LHL_list
    LLH[index]=LLH_list
    HHL[index]=HHL_list
    HLH[index]=HLH_list
    LHH[index]=LHH_list
    HHH[index]=HHH_list
    
    
    
#print(len(LLL[0]), len(HLL[0]), len(LHL[0]), size(LLH[0]),len(LLH[0]), HHL[0], HLH[0], LHH[0], HHH[0])
    
```

```python
def namestr(obj, namespace):
    return [name for name in namespace if namespace[name] is obj]
```

```python
import random
category=[LLL, HLL, LHL, LLH, HHL, HLH, LHH, HHH]
#select three points from the eight baskets, and save its pdb file.
print('Index', 'Traj','Bucket', 'size_bucket', 'ran1', 'ran2',\
      'frame1', 'frame1_min_his376_he3_dist','frame1_min_his376_asp372_dist','frame1_min_asp372_he3_dist',\
      'frame2', 'frame2_min_his376_he3_dist','frame2_min_his376_asp372_dist','frame2_min_asp372_he3_dist')
for index, traj in enumerate(trajs):
    for i, bucket in enumerate(category):
        ##bucket need to contain at least 2 samples to be seleted 
        if(len(bucket[index]) >= 2):
            ##random number: choose randomly 2 samples from the eight baskets
            option1, option2 = random.sample(range(len(bucket[index])), 2)
            print(index,namestr(traj, globals())[0], namestr(bucket, globals())[0], \
                  len(bucket[index]),option1,option2,\
                  bucket[index][option1][0], \
                  bucket[index][option1][1][0], bucket[index][option1][2][0],bucket[index][option1][3][0],\
                  bucket[index][option2][0], \
                  bucket[index][option2][1][0], bucket[index][option2][2][0],bucket[index][option2][3][0]\
                 )
            ##the two frames for each trajectory at each basket that are chosen
            n_frame1=bucket[index][option1][0]
            n_frame2=bucket[index][option2][0]
            #print(n_frame1, n_frame2)
            #save the chosen snapshots
            traj[n_frame1].save_pdb\
            ('/Users/caixiuhong/Dropbox/cai/btype_cco/snapshot/Btype_CcO_MD/more_md_pdbs/'+ \
            '%s_%s_%dns.pdb' % \
             (namestr(traj, globals())[0], namestr(bucket, globals())[0], n_frame1/10)) 
            traj[n_frame2].save_pdb\
            ('/Users/caixiuhong/Dropbox/cai/btype_cco/snapshot/Btype_CcO_MD/more_md_pdbs/'+ \
            '%s_%s_%dns.pdb' % \
             (namestr(traj, globals())[0], namestr(bucket, globals())[0], n_frame2/10)) 
```
