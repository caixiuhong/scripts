## 2019/04/09 plot crg of pls vs redox state for each structures at one graph

```python
#import data files
file1 = pandas.read_csv(r"C:\Users\PC\Dropbox\cai\btype_cco\crg_data\quick_run_mdlip\more_md_pdbs\crg_detail_ba3_more_md_pdbs.csv")
file2 = pandas.read_csv(r'C:\Users\PC\Dropbox\cai\btype_cco\crg_data\quick_run_mdlip\free_no_water\crg_detail_ba3.csv')
file3 = pandas.read_csv(r'C:\Users\PC\Dropbox\cai\btype_cco\crg_data\quick_run_mdlip\free_no_water\crg_detail_ba3_xray.csv')
```

```python
print(file1.shape)
print(file2.shape)
print(file3.shape)
```

```python
#calculate sum of charges at PLS for each substate and each structures
file1['sum_pls']=file1.iloc[:,1:17].sum(axis=1)
file2['sum_pls']=file2.iloc[:,1:17].sum(axis=1)
file3['sum_pls']=file3.iloc[:,1:17].sum(axis=1)
```

```python
#reshape the sum_pls, make 12 data one sturctue
n1=file1['sum_pls'].size
n2=file2['sum_pls'].size
n3=file3['sum_pls'].size
data1=file1['sum_pls'].values.reshape([n1//12,12])
data2=file2['sum_pls'].values.reshape([n2//12,12])
data3=file3['sum_pls'].values.reshape([n3//12,12])
```

```python
#categorize each strucs as [xry: bule, p1/dddro : orange, p2/pddro : red, p3/ppdro: green, p4/dpdro: cyan ]
states=[file1['state_snap'], file2['state_snap'], file3['state_snap']]
color_map={"p10": "orange", "dddro": "orange", "p20": "red", "pddro": "red", "p30": "green", "ppdro": "green",\
           "p40": "cyan","dpdro": "cyan"}
colors=[]
for i in range(len(states)):
    color=[]
    for j in range(0,len(states[i]), 12):
        for key, value in color_map.items(): 
            if key in states[i][j]:
                color.append(value)
                break
        if all(key not in states[i][j] for key in color_map):
            color.append("blue")
    colors.append(color)

print(colors)
```

```python
#plot the pls crg set
#categorize each strucs as [xry: bule, p1/dddro : orange, p2/pddro : red, p3/ppdro: green, p4/dpdro: cyan ]

outputpathway=r"C:\Users\PC\Dropbox\cai\btype_cco\crg_data\quick_run_mdlip"
#marker= ['^','o','v','X','s','*','<']
x_label=['r5','f1','f3','f5','o1','o3','o5','e1','e3','e5','r1', 'r3']

plt.figure()
plt.figure(figsize=(15, 5))
plt.xticks(list(range(12)),x_label)
data=[data1,data2, data3]
for i,datai in enumerate(data):
    for j, struc in enumerate(datai):     
        plt.plot(struc,label=str(i)+str(j), alpha=0.8, color=colors[i][j])
        #plt.legend()
        #plt.ylim(-5.0,2.0)
plt.title('PLS vs State')
plt.xlabel('substate')
plt.ylabel('charge')
plt.savefig(outputpathway+'/'+'All_Crg_State_ba3.png', dpi=300)
plt.show()

#plt.clf()

#categorize each strucs as [xry: bule, p1/dddro : orange, p2/pddro : red, p3/ppdro: green, p4/dpdro: cyan ]
```

### Notes: 
Crystal structures all perform loaded property. They prefered to have **protonaed H376 and D372 and deprotonaed PRA3**.<br/>
Traj3(PRA protonated) and Traj2(PRA protonated) shows their pls is at intermediate state. Moreover, The PLS pka of Traj3 is higher than of Traj2. Traj3 structures have **H376** to change its pKa along with redox state. ** Changeable D372 and  PRA3** seems couple together to contribute neutral environment for H376. D372 is more likely to get protonated than PRA3.<br/> 
Traj2 structures prefer to have **D372** change its pKa and **deprotonated H376 and protonaed PRA3**.<br/>
Traj1(PRA deprotonated) perfers unloaded state for pls, and the ionization states with more lower charge (-5) can be selected for the Botlzman microstate distribution, especially when snapshot is titrated at f3 state. <br/>
Traj4 (PRA depronated) tends to perform like Traj2 or Traj1. <br/>
D287 have a stable pKa for every structures. E126B and PRD3 has quite stable pka too.<br/>
